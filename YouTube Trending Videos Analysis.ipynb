```python
def process_and_plot_videos(df):
    # Ensure 'publish_time' is in datetime format
    df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')  # Convert to datetime

    # Extract the year
    df['year'] = df['publish_time'].dt.year  # Extract the year

    # Count videos by year
    cdf = df['year'].value_counts().reset_index()
    cdf.columns = ['year', 'No_of_videos']

    # Print the cdf for verification
    print(cdf)

    # Create the bar plot
    fig, ax = plt.subplots()

    # Specify the order of years to plot
    sns.barplot(x='year', y='No_of_videos', data=cdf, 
                palette=sns.color_palette(['#ff764a', '#ffa600'], n_colors=7), 
                ax=ax, order=cdf['year'].astype(str).tolist())

    # Set x and y labels
    ax.set(xlabel='Year', ylabel='No. of videos')

    # Show the plot
    plt.show()

# Usage
# Assuming df is your DataFrame containing the data
# Call the function to process and plot
process_and_plot_videos(df)
```

# Do the trending video titles contain captitalized words?

Now we want to see how many trending video titles contain at least a capitalized word (e.g. HOW). To do that, we will add a new variable (column) to the dataset whose value is True if the video title has at least a capitalized word in it, and False otherwise.

```python
def contains_capitalized_word(s):
    for w in s.split():
        if w.isupper():
            return True
    return False


df["contains_capitalized"] = df["title"].apply(contains_capitalized_word)

value_counts = df["contains_capitalized"].value_counts().to_dict()
fig, ax = plt.subplots()
_ = ax.pie([value_counts[False], value_counts[True]], labels=['No', 'Yes'], 
           colors=[PLOT_COLORS[4], '#ffa600'], textprops={'color': '#040204'}, startangle=45)
_ = ax.axis('equal')
_ = ax.set_title('Capitalized Word?')
```

```python
df["contains_capitalized"].value_counts(normalize=True)
```

We can see that around `40%` of trending video titles contain at least a capitalized word. We will later use this added new column contains_capitalized in analyzing correlation between variables.

# Trending videos and their publishing time



add two columns to represent the date and hour of publishing each video, then delete the original publish_time column because we will not need it anymore.

```python
df["publishing_day"] = df["publish_time"].apply(
    lambda x: datetime.datetime.strptime(x[:10], "%Y-%m-%d").date().strftime('%a'))
df["publishing_hour"] = df["publish_time"].apply(lambda x: x[11:13])
df.drop(labels='publish_time', axis=1, inplace=True)
```

```python
df.head()
```

Now we can see which days of the week had the largest numbers of trending videos.

```python
# Assuming df is already defined and contains the 'publishing_day' column
cdf = df["publishing_day"].value_counts()\
        .to_frame().reset_index().rename(columns={"index": "publishing_day", "publishing_day": "No_of_videos"})

# Print the DataFrame to verify column names
print(cdf.head())  # Check the first few rows of cdf

# Correct the renaming
cdf.columns = ['publishing_day', 'No_of_videos']  # Ensure the columns are named correctly

fig, ax = plt.subplots()

# Use the correct column names
_ = sns.barplot(x="publishing_day", y="No_of_videos", data=cdf, 
                palette=sns.color_palette(['#003f5c', '#374c80', '#7a5195', 
                                           '#bc5090', '#ef5675', '#ff764a', '#ffa600'], n_colors=7), ax=ax)

_ = ax.set(xlabel="Publishing Day", ylabel="No. of videos")

# Show the plot
plt.show()
```

We can see that the number of trending videos published on Sundayis noticeably less than the number of trending videos published on other days of the week.

use publishing_hour column to see which publishing hours had the largest number of trending videos.

```python
# Assuming df is already defined and contains the 'publishing_hour' column
cdf = df["publishing_hour"].value_counts().to_frame().reset_index()\
        .rename(columns={"index": "hour", "publishing_hour": "No_of_videos"})

# Print the DataFrame to verify column names
print(cdf.head())  # Check the first few rows of cdf

# Ensure the columns are named correctly
cdf.columns = ['publishing_hour', 'No_of_videos']  # Set the correct column names

# Convert 'publishing_hour' to a categorical type with a specified order
cdf['publishing_hour'] = pd.Categorical(cdf['publishing_hour'], 
                                        categories=[f"{i:02d}" for i in range(24)], 
                                        ordered=True)

fig, ax = plt.subplots()

# Use the correct column names
_ = sns.barplot(x="publishing_hour", y="No_of_videos", data=cdf, 
                palette=sns.cubehelix_palette(n_colors=24), ax=ax)

_ = ax.set(xlabel="Publishing Hour", ylabel="No. of videos")

# Show the plot
plt.show()
```

We can see that the period between 11AM(4.30 PM in India) and 4PM(9.30 PM in India), peaking between 12PM(5.30 PM in India) and 2PM(7.30 PM in India), had the largest number of trending videos. We notice also that the period between 8PM(1.30 AM in India) and 9PM(2.30 AM in India) has the smallest number of trending videos.

# Conclusions

### Here are the some of the results we extracted from the analysis: 

* We analyzed a dataset that contains information about YouTube trending videos for 205 days. The dataset was collected in 2017 and 2018 and contains `37352` video entries.

* The category that has the largest number of trending videos is 'Entertainment' with `16,712` videos, followed by 'News & Politics' category with `5241` videos, followed by 'Music' category with `3858` videos.
